```python
#realizamos diagrama de caja o bigote de cada columna en base a los datos curados
fig= plt.figure(figsize=(15,8))
data.plot(kind='box', vert=False)
plt.title ("Valores atipicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Filtro por columnas no consecutivas, ya que no contamos con columnas contiguas de datos
datos=data.iloc[:,[13,14,16,21,22,29,31,33,41,42,43,44,45,49,51,56,57,58,59,60,78,90]] #Columnas no consecutivas
datos
```

```python
#corroboramos valores nulos
valores_nulos=datos.isnull().sum()
valores_nulos
```

```python
#DEBIDO AL CONTEXTO DE LA PREGUNTA, SE ASUME QUE CADA UNO DE LOS VALORES LLEGAN A SER 0 POR EL CONTEXTO DE LA PREGUNTA (no se cobra fee, no tienen refrijeradores, no hay competenicia, etc...)
datos=datos.fillna(0)
```

```python
#corroboramos valores nulos
valores_nulos=datos.isnull().sum()
valores_nulos
```

ELIMINACION DE **OUTLIERS**

```python
#en este punto se define la desviacion estandar en base a los datos restantes
y=data
limite_superior= y.mean()+3*y.std()
limite_inferior= y.mean()-3*y.std()
print("Limite superior permitido", limite_superior)
print("Limite inferior permitido", limite_superior)
```

```python
#encontramos outliers del dataframe curado
data1= data[(y>limite_superior)|(y<limite_inferior)]
data1
```

```python
#convertir outliers en nulos en dataframe
datos2 = data[(y<=limite_superior)&(y>=limite_inferior)]
datos2
```

```python
#corroboramos valores  nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#reemplazamos valores atipicos (nulos) con "mean"
datos3=data1.copy()
datos3=datos3.fillna(round(data1.mean(),1))
datos3
```

```python
#convertir dataframe a csv
datos3.to_csv("Datos_limpios_microretailer_mit_lift_lab.csv")
```
